```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
data = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
target_names = iris['target_names']

# Plotting the Iris dataset
plt.figure(figsize=(8, 6))

# Scatter plot for Sepal Length vs Sepal Width
plt.scatter(data[data['target'] == 0]['sepal length (cm)'], data[data['target'] == 0]['sepal width (cm)'],
            label=target_names[0], c='r', marker='o')
plt.scatter(data[data['target'] == 1]['sepal length (cm)'], data[data['target'] == 1]['sepal width (cm)'],
            label=target_names[1], c='g', marker='x')
plt.scatter(data[data['target'] == 2]['sepal length (cm)'], data[data['target'] == 2]['sepal width (cm)'],
            label=target_names[2], c='b', marker='s')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Sepal Length vs Sepal Width')
plt.legend(loc='upper right')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data.T  # Transpose the data to have 4 rows and 150 columns
species = iris.target  # Vector of length 150 with species (0, 1, or 2)
n = X.shape[1]

# Create a figure to plot pairwise relationships
plt.figure(1, figsize=(10, 10))

# Define species labels
species_labels = ['Setosa', 'Versicolor', 'Virginica']

# Define feature names
feature_names = iris.feature_names

# Plot x_i vs. x_j for all combinations of i, j
subplot_num = 1
for i in range(4):
    for j in range(i):
        plt.subplot(3, 2, subplot_num)  # Adjust the subplot parameters here
        for k in range(3):
            plt.scatter(X[i][species == k], X[j][species == k], s=7, label=species_labels[k])
        plt.xlabel(feature_names[i])
        plt.ylabel(feature_names[j])
        plt.legend()
        subplot_num += 1

plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
irisModule = load_iris()
dataset = np.array(irisModule.data)
```

```python
covarianceMatrix = pd.DataFrame(data = np.cov(dataset, rowvar = False), columns = irisModule.feature_names, index = irisModule.feature_names)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covarianceMatrix)
display(eigenvalues)
display(eigenvectors)
```

```python
totalSum = sum(eigenvalues)
variablesExplained = [(i / totalSum) for i in sorted(eigenvalues, reverse = True)]
display(variablesExplained)
```

```python
cumulative_explained_variance = np.cumsum(variablesExplained)


# Create a single chart with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the bar chart of explained variances
ax1.bar(range(1, len(variablesExplained) + 1), variablesExplained, color='b', alpha=0.7)
ax1.set_xlabel('Principal Component')
ax1.set_ylabel('Proportion of Variance Explained', color='b')
ax1.set_title('Variance Explained and Cumulative Explained Variance')
ax1.grid(True)

# Create a secondary y-axis for cumulative explained variance
ax2 = ax1.twinx()

# Plot the cumulative explained variance
ax2.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, color='r', marker='o')
ax2.set_ylabel('Cumulative Explained Variance', color='r')

plt.tight_layout()
plt.show()
```
